```python
class Product:
    def __init__(self, ID, name, category, total_stock, price_per_unit):
        self.ID = ID
        self.name = name
        self.category = category
        self.total_stock = total_stock
        self.price_per_unit = price_per_unit
    
    def modify(self, key, newvalue):
        if key=='ID':
            self.ID=newvalue
        elif key=='name':
            self.name=newvalue
        elif key=='category':
            self.category=newvalue
        elif key=='stock':
            self.total_stock=newvalue
        else:
            self.price_per_unit=newvalue
        
    def sell(self,quantity):
        self.total_stock=(self.total_stock-quantity)
        
   
products=[]
with open('/Users/mariapiabellini/Desktop/Unicatt/IT Coding/progetto/inventory_products_supermarket.txt', 'r') as file:
        for line in file:
            data = line.strip().split(',')
            product = Product(int(data[0]), data[1], data[2], int(data[3]),float(data[4]))
            products.append(product)
            
for i in range(0,len(products)):
    print(products[i].name)
```

```python
class Customer:
    def __init__(self, ID, f_name, l_name, address, balance):
        self.ID = ID
        self.f_name = f_name
        self.l_name = l_name
        self.address = address
        self.balance = balance
        
    def purchase(self, product, quantity):
        self.balance=(self.balance-(product.price_per_unit)*quantity)
        
    def checkBalance(self):
        print(self.balance)

   
customers=[]
with open('/Users/mariapiabellini/Desktop/Unicatt/IT Coding/progetto/customers_supermarket.txt', 'r') as file:
        for line in file:
            data = line.strip().split(',')
            customer = Customer(int(data[0]), data[1], data[2], data[3],float(data[4]))
            customers.append(customer)
            
for i in range(0,len(customers)):
    print(customers[i].f_name)
```

```python
import os

def save_to_file(customers, filename):
    with open(filename, 'w') as file:
        for customer in customers:
            file.write(f"{customer.ID},{customer.f_name},{customer.l_name},{customer.address},{customer.balance}\n")

def new_customer(ID, f_name, l_name, address, balance):
    new_cust=Customer(ID, str(f_name), str(l_name), str(address), float(balance))
    customers.append(new_cust)
    
def main():
    while True:
        print('Type "display" to display the list of customers in the system.')
        print('Type "add" to add a new customer to the system.')
        print('Type "delete" to delete a specified customer.')
        print('Type "save" to save the list to a file')
        print('Type "exit" to exit')
        
        choice=input('Enter your choice -> ')

        if choice=='display':
            for customer in customers:
                print('ID:',customer.ID,',','First name:', customer.f_name,',','Last name:', customer.l_name,',','Address:', customer.address,',','Balance:', customer.balance,'\n')
        
        elif choice=='add':
            print('please insert the new customer attributes \n')
            a=input('ID -> ')
            b=input('First name ->')
            c=input('Last name ->')
            d=input('Address ->')
            e=input('Balance ->')
            new_customer(a,b,c,d,e)
            
        elif choice == 'delete':
            choice2 = input('Insert the first name of the customer you want to delete: ')
            choice3 = input('Insert the last name of the customer you want to delete: ')

            for i in range(len(customers)):
                if customers[i].f_name == choice2 and customers[i].l_name == choice3:
                        customers.remove(customers[i])
                        
        elif choice=='save':
            save_to_file(customers, filename)
                        
        elif choice=='exit':
            break
        else:
            print('Invalid choice, please insert a valid choice \n')
if __name__ == "__main__":
    main()
```
